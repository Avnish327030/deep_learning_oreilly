```python
import torch
```

```python
from sklearn import datasets
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
n_pts =100
centers = [[-0.5,0.5],[0.5,-0.5]]
X,y = datasets.make_blobs(n_samples=n_pts,random_state=64,centers=centers, cluster_std=0.4)
print(X,y)
```

```python
def scatter_plot(X,y):
    plt.scatter(X[y==0,0],X[y==0,1])
    plt.scatter(X[y==1,0],X[y==1,1])
```

```python
x_data = torch.Tensor(X)
y_data = torch.Tensor(y).view(100,1)
```

### Perceptron based neural network

```python
test = np.linspace(-1,1)
```

```python
plt.plot(test,np.log2(test),"o")
```

```python
from torch import nn
```

```python
class Model(nn.Module):
    
    def __init__(self, input_size, output_size):
        super().__init__()
        self.linear = nn.Linear(in_features=input_size,out_features=output_size)
    
    def forward(self,X):
        pred = torch.sigmoid(self.linear(X))
        return pred
    
    def predict(self,X):
        prob = self.forward(X)
        prob = prob.item()
        if prob>0.5:
            return 1
        else:
            return 0
        
    
    
    
```

```python
torch.manual_seed(2)
model = Model(input_size=2,output_size=1)
```

```python
list(model.parameters())
```

```python
def get_param(model):
    [w,b] = model.parameters()
    w1,w2 = list(map(lambda x: x.item(),w.view(2)))
    b1 = b.item()
    print(f"w1: {w1} w2:{w2} and b1: {b1}")
    return w1,w2,b1
```

```python
print(get_param(model))
```

```python
def plot_fit(title,model,X,y):
    plt.title = title
    w1,w2,b1 = get_param(model)
    x1= np.array([-2.0,2.0])
    x2 = (w1*x1 + b1)/(-w2)
    plt.plot(x1,x2, "r")
    scatter_plot(X,y)
    
    
    
    
    
```

```python
plot_fit("Initial Model",model,X,y)
```

```python
criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs = 1000
```

```python
losses = []

for i in range(epochs):
    pred = model.forward(x_data)
    loss = criterion(pred, y_data)
    losses.append(loss.item())
    print(f"Epoch: {i} loss: {loss.item()}")
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plot_fit("Train Model",model,X,y)
```

```python
plt.plot(range(epochs),losses)
```

```python
point1 = torch.Tensor([[1.0,-1.0]])
point2 = torch.Tensor([[-1.0,1.0]])
plt.plot(point1.numpy()[0,0],point1.numpy()[0,1],"ro")
plt.plot(point2.numpy()[0,0],point2.numpy()[0,1],"ko")
print(f"Red point probability = {model.forward(point1).item()}")
print(f"Black point probability = {model.forward(point2).item()}")
print(f"Red point  = {model.predict(point1)}")
print(f"Black point  = {model.predict(point2)}")
```
